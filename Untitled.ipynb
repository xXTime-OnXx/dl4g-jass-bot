```python
import numpy as np
import matplotlib.pyplot as plt

# Time vector
time = np.linspace(0, 20, 200)

# Modified angle error to start at -0.4 and converge to zero
angle_error = (np.sin(time) * np.exp(-time / 5)) - 0.4 * np.exp(-time)

# Time step (delta_time) for the integral
delta_time = time[1] - time[0]
integral_error = np.cumsum(angle_error) * delta_time

# Plot the results
plt.figure(figsize=(10, 6))

# Plot 1: Winkelfehler über die Zeit
plt.subplot(2, 1, 1)
plt.plot(time, angle_error, label="Winkelfehler", color='b')
plt.axhline(0, color='black', linewidth=0.5)
plt.title("Winkelfehler über Zeit")
plt.ylabel("Winkelfehler (rad)")
plt.xlabel("Zeit (s)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt


time = np.linspace(0, 10, 100) 
kp = 0.5  
initial_angle_error = 30  



offset = 1 
angle_error_proportional_modified = initial_angle_error * np.exp(-kp * time) + offset


plt.figure(figsize=(8, 6))
plt.plot(time, angle_error_proportional_modified, label='Proportionalregler', color='blue', linewidth=2)
plt.axhline(1, color='red', linestyle='--', linewidth=1, label='Restfehler')  # Linie bei y=1 für Restfehler
plt.title("Winkelfehler bei reinem Proportionalregler")
plt.xlabel("Zeit (s)")
plt.ylabel("Winkelfehler (Grad)")
plt.grid(True)
plt.legend()


plt.show()
```
